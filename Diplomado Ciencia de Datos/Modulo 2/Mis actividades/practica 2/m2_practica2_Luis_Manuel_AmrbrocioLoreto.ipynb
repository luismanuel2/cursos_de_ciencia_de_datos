```python
def clean_text(text, pattern="[^a-zA-Z0-9]"):
    if isinstance(text, str):
        cleaned_text = unicodedata.normalize('NFD', text).encode('ascii', 'ignore')
        cleaned_text = re.sub(pattern, " ", cleaned_text.decode("ascii"), flags=re.UNICODE)
        cleaned_text = u' '.join(cleaned_text.lower().lstrip().strip().split())
    else:
        cleaned_text = text
    return cleaned_text
def eliminar_caracteres_especiales(texto):
    return re.sub(r'[^a-zA-Z0-9áéíóúüÜÁÉÍÓÚñÑ\s():\[\]/,?¡¿!\']', '', texto)
def contiene_url(mensaje):
    return bool(re.search(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', mensaje))
    nltk.download('stopwords')

def eliminar_stop_words(texto):
    palabras = texto.split()
    palabras_sin_stopwords = [palabra for palabra in palabras if palabra.lower() not in stop_words]
    return ' '.join(palabras_sin_stopwords)
def print_top_ngram(ngram):
    ngrams_individuo = {'Michael': [], 'Lm': []}
    for index, row in df.iterrows():
        target = row['target']
        mensaje = row['mensaje_ssw']
        ngram_tokens = mensaje.split()
        ngrams_list = list(ngrams(ngram_tokens, ngram))
        ngrams_individuo['Lm' if target == 1 else 'Michael'].extend([" ".join(ngram) for ngram in ngrams_list])
    
    
    top_ngrams = {'Michael': [], 'Lm': []}
    top_ngram_count = 10
    
    for individuo in ngrams_individuo:
        ngram_counts = Counter(ngrams_individuo[individuo])
        top_ngrams[individuo] = ngram_counts.most_common(top_ngram_count)
    for ngram_length in range(1, 6):  
        for individuo in top_ngrams:
            ngram_data = [ngram for ngram in top_ngrams[individuo] if len(ngram) > 0 and len(ngram[0].split()) == ngram_length]
            if not ngram_data:
                continue  # Ignora si no hay n-gramas disponibles
            ngram_labels, ngram_counts = zip(*ngram_data)
            plt.barh(ngram_labels, ngram_counts)
            plt.xlabel('Frecuencia')
            plt.title(f'N-gramas más comunes de longitud {ngram_length} para {individuo}')
            plt.gca().invert_yaxis()
            plt.show()
def contar_acentos_y_puntuacion(texto):
    # Utiliza una expresión regular para encontrar tanto acentos como signos de puntuación
    acentos_y_puntuacion = re.findall(r'[.,;!?áéíóúÁÉÍÓÚ]', texto)
    total_acentos_y_puntuacion = len(acentos_y_puntuacion)
    total_palabras = len(texto.split())  # Dividir el texto en palabras por espacios en blanco

    # Calcula el porcentaje de acentos y signos de puntuación en el texto
    porcentaje = (total_acentos_y_puntuacion / total_palabras) * 100

    return porcentaje


def porcentaje_errores_ortograficos(texto, idioma="es"):
    # Crea un diccionario para el idioma especificado (español por defecto)
    spell_checker = enchant.Dict(idioma)

    # Divide el texto en palabras
    palabras = texto.split()

    # Encuentra las palabras que contienen errores
    errores = [palabra for palabra in palabras if not spell_checker.check(palabra)]

    # Calcula el porcentaje de errores en el texto
    total_palabras = len(palabras)
    total_errores = len(errores)
    porcentaje = (total_errores / total_palabras) * 100

    return porcentaje


def get_cv_scores_report_classification(estimator, X, y, n_splits):
    cv_scores = cross_validate(
                    estimator = estimator,
                    X         = X,
                    y         = y,
                    scoring   = {
                                'accuracy': make_scorer(accuracy_score), 
                                'recall': make_scorer(recall_score, average='weighted'), 
                                'roc_auc_ovr': 'roc_auc_ovr'
                                },
                    cv        = RepeatedStratifiedKFold(n_splits=n_splits, n_repeats=5, random_state=333),
                )
    
    # Convertir el diccionario a dataframe para facilitar la visualización
    cv_scores = pd.DataFrame(cv_scores)
    print(f"Accuracy en CV: mean {cv_scores.test_accuracy.mean():.2f}, std {cv_scores.test_accuracy.std():.2f}")
    print(f"Recall en CV: mean {cv_scores.test_recall.mean():.2f}, std {cv_scores.test_recall.std():.2f}")
```

# Dataset

el dataset corresponde a una conversacion extraida de Whatsapp entre yo (Lm) y un amigo llamado Michael

```python
archivo = open("_chat.txt", "r")

for i in range(5):
    linea = archivo.readline()
    if not linea:
        break  
    print(linea, end="") 
archivo.close()
```

la estructura de la conversacion es

{Fecha}, {Hora} - {Autor}: {Mensaje}

el objetivo será predecir quien de los dos envio el mensaje, primero pasemos el texto a un dataset, utilizaremos 1 para cuando el mensaje fue mio y 0 para cuando fue de Michael


```python
import pandas as pd

data = []

with open("_chat.txt", "r", encoding="utf-8") as archivo:
    for linea in archivo:
        linea_mod =  eliminar_caracteres_especiales(linea) #unidecode(linea)#linea.encode('latin1','ignore').decode('utf-8')
        partes = linea_mod.strip().split("] ")
        fecha_hora = partes[0].strip("[").split(", ")
        if len(partes) >= 2:
            remitente = partes[1].split(": ")[0]
            mensaje = " ".join(partes[1].split(": ")[1:])
            
            if len(fecha_hora) >= 2 :
                if remitente in {'Lm','Michael'}:
                    data.append([fecha_hora[0], fecha_hora[1], remitente, mensaje])

df = pd.DataFrame(data, columns=["fecha", "hora", "remitente", "mensaje"])

print(df.head())
```

# Modelo

```python
cv = TfidfVectorizer(ngram_range=(1, 3), 
                     min_df=10, 
                     max_features=100)
cv.fit(df["mensaje_ssw"])
```

```python
ls_features = ['porcentaje_Errores','porcentaje_Puntuacion','polaridad','num_palabras']
X= df[ls_features].fillna(0).merge(pd.DataFrame(data = cv.transform(df["mensaje_ssw"]).todense(), columns = cv.get_feature_names_out(),index = df.index), left_index=True, right_index=True, how='inner')
y = df['target']
```

```python
kb = SelectKBest(k=30, score_func=f_classif)
kb.fit(X, y)
```

```python
ls_best = [x for x, y in zip(X.columns, kb.get_support()) if y]
X = X[ls_best]
```

```python
ls_best
```

```python


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
space = {
    'model_type': hp.choice('model_type', [
        {
            'type': 'gaussian',
            'var_smoothing': hp.loguniform('gaussian_var_smoothing', -10, -6)
        },
        {
            'type': 'multinomial',
            'alpha': hp.uniform('multinomial_alpha', 0, 2) 
        },
        {
            'type': 'bernoulli',
            'alpha': hp.uniform('bernoulli_alpha', 0, 2), 
            'binarize': hp.uniform('bernoulli_binarize', 0, 1)  # Uniforme entre 0 y 1
        }
    ])
}
```

```python
def objective(params):
    if params['model_type']['type'] == 'gaussian':
        clf = GaussianNB(var_smoothing=params['model_type']['var_smoothing'])
    elif params['model_type']['type'] == 'multinomial':
        clf = MultinomialNB(alpha=params['model_type']['alpha'])
    else:  # bernoulli
        clf = BernoulliNB(alpha=params['model_type']['alpha'], binarize=params['model_type']['binarize'])
    
    # Asumiendo X_train e y_train son tus datos
    score = cross_val_score(clf, X_train, y_train, cv=5).mean()
    return {'loss': -score, 'status': STATUS_OK}

trials = Trials()
best = fmin(fn=objective, space=space, algo=tpe.suggest, max_evals=100, trials=trials)
```

```python
if best['model_type'] == 0:  # Gaussian
    nb = GaussianNB(var_smoothing=best['gaussian_var_smoothing'])
elif best['model_type'] == 1:  # Multinomial
    nb = MultinomialNB(alpha=best['multinomial_alpha'])
else:  # Bernoulli
    nb = BernoulliNB(alpha=best['bernoulli_alpha'], binarize=best['bernoulli_binarize'])
    
nb.fit(X_train, y_train)
```

```python
def pred_text(txt):
    polaridad  = sentiment.sentiment(txt)
    pct_errores = porcentaje_errores_ortograficos(txt)
    pct_signos = contar_acentos_y_puntuacion(txt)
    num_palabras = len(txt.split(' '))
    df_pred = pd.DataFrame({'mensaje_ssw':[eliminar_stop_words(txt).lower()],'porcentaje_Puntuacion':[pct_signos],'polaridad':[polaridad],
                            'porcentaje_Errores':[pct_errores],'num_palabras':[num_palabras]})
    df_pred = df_pred.fillna(0).merge(pd.DataFrame(data = cv.transform(df_pred["mensaje_ssw"]).todense(), columns = cv.get_feature_names_out(),index = df_pred.index), left_index=True, right_index=True, how='inner')
    df_pred = df_pred[ls_best]
    sal = ''
    if nb.predict(df_pred)  == 0:
        sal = 'Michael'
    else: 
        sal  ='Lm'
    return sal
```
